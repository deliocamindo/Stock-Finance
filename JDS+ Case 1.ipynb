```python
# Create a list with the stocks we want to download
stocks = ['TSLA', 'NIO', 'IQQH.F', 'BTC-USD', 'ETH-USD', 'LTC-USD', 
          'AMZN', 'TWTR', 'FB', 'SQ', 'PYPL', 'BRK-A',
          'CSPX.AS', 'GC=F', 'SI=F', 'CL=F', 'UA', 'IT']
```

```python
base = datetime.date.today()

date_list = [base - datetime.timedelta(days=x) for x in range(1)]
date_list.sort()
# Create a empty DataFrame that i can use to add more columns
database = pd.DataFrame(index = date_list)

# This for loop adds each assets one by one
for i in stocks:
    i = yf.Ticker(i)
    i = i.history(period='max')[['Open']] 
    database = database.merge(i, how ='outer', left_index=True, right_index=True)
    
database.columns = stocks
```

###### Cleaning the dataset

Change the NaN values for the mean of the column



```python
# Compute the mean of each column

mean_nio = database['NIO'].mean()

mean_iqqhf = database['IQQH.F'].mean()

mean_btc_usd = database['BTC-USD'].mean()

mean_eth_usd = database['ETH-USD'].mean()

mean_ltc_usd = database['LTC-USD'].mean()

mean_twtr = database['TWTR'].mean()

mean_fb = database['FB'].mean()

mean_sq = database['SQ'].mean()

mean_pypl = database['PYPL'].mean()

mean_cspxas = database['CSPX.AS'].mean()

mean_gcf = database['GC=F'].mean()

mean_sif = database['SI=F'].mean()

mean_clf = database['CL=F'].mean()

mean_ua = database['UA'].mean()
```

###### Change the NaN value of each column

```python
database['NIO'].fillna(mean_nio, inplace=True)

database['IQQH.F'].fillna(mean_iqqhf, inplace=True)

database['BTC-USD'].fillna(mean_btc_usd, inplace=True)

database['ETH-USD'].fillna(mean_eth_usd, inplace=True)

database['LTC-USD'].fillna(mean_ltc_usd, inplace=True)

database['TWTR'].fillna(mean_twtr, inplace=True)

database['FB'].fillna(mean_fb, inplace=True)

database['SQ'].fillna(mean_sq, inplace=True)

database['PYPL'].fillna(mean_pypl, inplace=True)

database['CSPX.AS'].fillna(mean_cspxas, inplace=True)

database['GC=F'].fillna(mean_gcf, inplace=True)

database['SI=F'].fillna(mean_sif, inplace=True)

database['CL=F'].fillna(mean_clf, inplace=True)

database['UA'].fillna(mean_ua, inplace=True)
```

#### Normalization or Standardzation

Testing the multiple methods

```python
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

###### METHOD#1: MIN-MAX SCALER
This is the simpe way tonormalize data


```python
min_max_scaler = MinMaxScaler()
pd.DataFrame(min_max_scaler.fit_transform(database[['TSLA', 'NIO', 'SQ', 'IT']])).tail(2000).plot()
```

#### METHOD#2 MAX-ABS SCALER

```python
max_abs_scaler = MaxAbsScaler()
pd.DataFrame(max_abs_scaler.fit_transform(database[['TSLA', 'NIO', 'SQ', 'IT']])).tail(2000).plot()
```

#### METHOD#3  STANDARDSCALER

```python
standard_scaler = StandardScaler()
pd.DataFrame(standard_scaler.fit_transform(database[['TSLA', 'NIO', 'FB', 'IT']])).tail(2000).plot()
```

##### METHOD#4 ROBUSTSCALER

```python
Robust_scaler = RobustScaler()
pd.DataFrame(Robust_scaler.fit_transform(database[['TSLA', 'NIO', 'FB', 'IT']])).tail(2000).plot()
```

```python
base = datetime.date.today()
date_list = [base - datetime.timedelta(days=x) for x in range(1)]
date_list.sort()
porfolio = pd.DataFrame(index = date_list)

selected_assets = ['CL=F', 'LTC-USD', 'SI=F', 'TWTR']

for i in selected_assets:
    f = yf.Ticker(i)
    f = f.history(period='max')
    f = f[['Open']]
    f = f/float(f.loc[['2019-07-01']]['Open'])
    porfolio = porfolio.merge(f, how = 'outer', left_index =True, right_index=True)

porfolio.columns = selected_assets

for i in selected_assets:
    porfolio = porfolio[porfolio[i].notna()]
```

```python
porfolio.tail(1200).reset_index(drop = True).plot()
```
